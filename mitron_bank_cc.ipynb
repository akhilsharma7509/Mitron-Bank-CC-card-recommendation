Analyzing accorindg to customer demographics

```python
mitron["age_group"].value_counts()
```

```python
sns.countplot(x='age_group', data=mitron)
plt.title('Customer Segmentation by Age Group')
plt.show()
```

```python
mitron["category"].value_counts()
```

```python
total_spend_category=mitron.groupby("category")["spend"].sum().sort_values(ascending=False)
total_spend_category
```

```python
plt.figure(figsize=(15,6))
plt.bar(total_spend_category.index,total_spend_category)
```

so most of the money spended is on the bill  followed  by groceries and others

```python
gender_spend=mitron.groupby("gender")["spend"].mean()
sns.barplot(x=gender_spend.index,y=gender_spend.values)
plt.title("average spend based on gender")
```

```python
spend_mode= mitron.groupby("payment_type")["spend"].mean()
spend_mode
```

```python
sns.barplot(x=spend_mode.index,y=spend_mode.values)
plt.title("average spend by payment type")
```

so most of the payment is done through credit card followed by upi and others


```python
plt.figure(figsize=(8,8))
sns.boxplot(x="age_group",y="spend",data=mitron)
plt.title('Average Spending by Age Group')
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.pointplot(x='city', y='spend', data=mitron)
plt.title('Average Spending by City')
plt.show()
```

most of the money people spend on mumbai city followed by delhi ncr and least at chennai city

```python
mitron["income_utilization"]=mitron["spend"]/mitron["avg_income"]
mitron.groupby("customer_id")["income_utilization"].mean()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(mitron["avg_income"],bins=20,kde=True)
plt.xlabel("income_utilization")
plt.ylabel("Frequency")
```

prople average earning between 35k-45k spend more  in comparison with 60k to 70k

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='occupation', y='spend', data=mitron, estimator='mean', errwidth=0)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='occupation', y='spend', hue='category', data=mitron, estimator='mean', errwidth=0)
plt.title('Spending Patterns by Category and Occupation')
plt.xlabel('occupation')
plt.ylabel('Average Spending')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Occupation', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='age_group', y='spend', hue='occupation', data=mitron, estimator='mean', errwidth=0)
plt.title('Spending Patterns by Category and Occupation')
plt.xlabel('age_group')
plt.ylabel('Average Spend')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Occupation', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='category', y='spend', hue='gender', data=mitron, estimator='mean', errwidth=0)
plt.title('Spending Patterns by Category and gender')
plt.xlabel('category')
plt.ylabel('Average Spend')
plt.xticks(rotation=45, ha='right')
plt.legend(title='gender', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='month', y='spend', data=mitron,estimator="sum", errwidth=0)
plt.title('Spending Patterns by Category and gender')
plt.xlabel('month')
plt.ylabel(' Spend freuqency')
plt.xticks(rotation=45, ha='right')
plt.show()
```
